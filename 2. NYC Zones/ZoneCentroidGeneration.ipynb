```python
centroid_data = pd.DataFrame(columns=["Zone", "Latitude", "Longitude"])

with open('NYC Taxi Zones.geojson', 'r') as f:
    data = json.load(f)

for feature in data['features']:
    zone = feature['properties']['zone']
    geom = shape(feature['geometry'])

    if geom.geom_type == 'MultiPolygon':
        centroid = geom.geoms[0].centroid  
    elif geom.geom_type == 'Polygon':
        centroid = geom.centroid

    longitude, latitude = centroid.x, centroid.y

    if zone not in centroid_data['Zone'].values:
        # Append new row to the DataFrame
        centroid_data = centroid_data.append({
            "Zone": zone,
            "Latitude": latitude,
            "Longitude": longitude
        }, ignore_index=True)

# Print the DataFrame to verify results
centroid_data
```

```python
result_df = pd.merge(zone_data, centroid_data, on='Zone', how='inner')  # Try changing 'how' to 'outer', 'left', 'right'
pd.set_option('display.max_rows', None)
result_df.head(300)
    
```
