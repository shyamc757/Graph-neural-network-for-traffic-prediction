```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
```

```python
monthly_aggregated_df = pd.DataFrame()
# temp_df = pd.DataFrame()

for year in range(2013, 2024):
    print(f"I started year {year}")

    df = pd.read_csv(f"ProcessedCitibikeData/ProcessedCitibikeData/CitibikeProcessed{year}.csv")
    
    df = df[~((df['start_zone_encoded'] == 264) | (df['start_zone_encoded'] == 265) | (df['end_zone_encoded'] == 264) | (df['end_zone_encoded'] == 265))]
    df['start_time'] = pd.to_datetime(df['unix_start_time'], unit='s')
    df['stop_time'] = pd.to_datetime(df['unix_stop_time'], unit='s')
    df['month'] = df['start_time'].dt.month
    df = df[['start_zone_encoded', 'end_zone_encoded', 'user_member_encoded', 'start_time', 'stop_time', 'month']]
    df['weight'] = 0
    
#     df = pd.concat([temp_df, df], ignore_index=True)
#     temp_df = pd.DataFrame()

    for month in range(1, 13):
        monthly_df = df[(df['month'] == month)]
        
        total_df = monthly_df.groupby(['start_zone_encoded', 'end_zone_encoded', 'user_member_encoded']).agg({'start_time':'min', 'stop_time':'max', 'weight':'count'}).reset_index()
        monthly_aggregated_df = pd.concat([monthly_aggregated_df, total_df], ignore_index=True)
```

```python
monthly_aggregated_df['start_time'] = monthly_aggregated_df['start_time'].astype('int64') // 10**9
monthly_aggregated_df['stop_time'] = monthly_aggregated_df['stop_time'].astype('int64') // 10**9
```

```python
monthly_aggregated_df.to_csv('monthlyAggregatedCitibike.csv', index=False)
```
